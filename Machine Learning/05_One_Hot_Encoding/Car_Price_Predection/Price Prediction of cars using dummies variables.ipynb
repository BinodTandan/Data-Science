```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel("Mileage(Miles)")
plt.ylabel("Sell Price($)")
plt.scatter(df.Mileage,df['Sell Price($)'], color='red', marker='*')
```

## Using Pandas to create Dummy Variables

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merge = pd.concat([df,dummies],axis='columns')
merge
```

```python
final = merge.drop(['Car Model','Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.predict([[50000,5,0,0]])
```

```python
model.predict([[40000,4,0,1]])
```

## Accuracy of the model

```python
model.score(X,y)
```

## Predict Price of Mercedez Benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])
```

## Predict Price of BMW X5  that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```
